## Factory production planning

A company has evaluated the profit from two differents products A and B. The factory has two workshops to manufacture the products. Considering the data provided, how many products of type A and B have to be produced in order to maximize the profit ?

For exemple, workshop 1 needs 4 hours to manufacture one product A. The workshop 1 cannot be used more than 800 hours.

|               | Workshop 1    | Workshop 2   | Profit |
| ------------- |:-------------:| :-----------:|:-----------:|
| Product A     | 4   hours     | 4 hours      | 50 € |
| Product B     | 10 hours      |   2 hours    | 75 € |
| Max. Capacity | 800 hours      |    320 |           |


## Optimisation problem

The production planning can expressed as a linear optimisation problem.

$$ \max{  Z= 50 x_A+ 75 x_B} $$
$$ s.c. $$
$$ 4x_A+10x_B \leq 800 $$
$$4x_A+2x_B \leq 320$$
$$x_A, x_B \geq 0$$

$x_A$ is the number for products A to manufacture.

$x_B$ is the number for products B to manufacture.

$50 x_A+ 75 x_B$ is the total profit.
